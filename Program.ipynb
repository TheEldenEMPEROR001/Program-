```python
import mysql.connector as sqltor

# --- Connect to MySQL ---
mycon = sqltor.connect (host="192.168.0.3", user="SHS24BANN39", passwd="SHS24BANN39", database="SHS24BANN39_")
cursor = mycon.cursor()

# --- Function to list all available cases ---
def show_cases():
    cursor.execute("SELECT case_id, case_title FROM cases")
    cases = cursor.fetchall()
    
    print("\nAvailable Cases:")
    for case_id, title in cases:
        print(f"  {case_id}. {title}")
    
    try:
        choice = int(input("\nEnter the case number you want to play: "))
        play_case(choice)
    except:
        print("Invalid choice.")

# --- Function to play a selected case ---
def play_case(case_id):
    # Fetch case info
    cursor.execute("SELECT case_title, case_description FROM cases WHERE case_id=%s", (case_id,))
    case = cursor.fetchone()
    
    if not case:
        print("Case not found.")
        return
    
    case_title, case_description = case
    print(f"\nCASE {case_id}: {case_title}")
    print(case_description)
    print("-" * 60)
    
    # Fetch questions and options using JOINs
    cursor.execute("""
        SELECT 
            q.question_id,
            q.question_text,
            a.answer_id,
            a.option_text,
            a.is_correct
        FROM questions q
        JOIN answers a ON q.question_id = a.question_id
        WHERE q.case_id = %s
        ORDER BY q.question_id, a.answer_id
    """, (case_id,))
    
    rows = cursor.fetchall()
    if not rows:
        print("No questions found for this case.")
        return
    
    # Group data by question
    questions_dict = {}
    for q_id, q_text, a_id, opt_text, is_correct in rows:
        if q_id not in questions_dict:
            questions_dict[q_id] = {
                "question_text": q_text,
                "options": []
            }
        questions_dict[q_id]["options"].append((a_id, opt_text, is_correct))
    
    # Gameplay
    score = 0
    for q_id, data in questions_dict.items():
        print(f"\n{data['question_text']}")
        options = data["options"]
        
        for i, (_, opt_text, _) in enumerate(options, start=1):
            print(f"  {i}. {opt_text}")
        
        try:
            choice = int(input("Choose your answer (1-4): "))
            if choice < 1 or choice > len(options):
                raise ValueError
        except ValueError:
            print("Invalid input, skipping question.")
            continue
        
        chosen = options[choice - 1]
        if chosen[2] == 1:
            print("Correct!")
            score += 1
        else:
            print("Wrong.")
    
    print("-" * 60)
    print(f"Final Score for Case {case_id}: {score}/{len(questions_dict)}")
    print("=" * 60)

# --- Main Menu ---
def main():
    print("=================================")
    print(" Welcome to the Detective Game ")
    print("=================================")
    show_cases()


# Run the game
def main():
    cursor.execute("SELECT case_id, case_title FROM cases")
    cases = cursor.fetchall()
    
    print("\nAvailable Cases:")
    for case_id, title in cases:
        print(f"  {case_id}. {title}")
    
    try:
        choice = int(input("\nEnter the case number you want to play: "))
        play_case(choice)
    except:
        print("Invalid choice.")

# --- Function to play a selected case ---
def play_case(case_id):
    # Fetch case info
    cursor.execute("SELECT case_title, case_description FROM cases WHERE case_id=%s", (case_id,))
    case = cursor.fetchone()
    
    if not case:
        print("Case not found.")
        return
    
    case_title, case_description = case
    print(f"\nCASE {case_id}: {case_title}")
    print(case_description)
    print("-" * 60)
    
    # Fetch questions and options using JOINs
    cursor.execute("""
        SELECT 
            q.question_id,
            q.question_text,
            a.answer_id,
            a.option_text,
            a.is_correct
        FROM questions q
        JOIN answers a ON q.question_id = a.question_id
        WHERE q.case_id = %s
        ORDER BY q.question_id, a.answer_id
    """, (case_id,))
    
    rows = cursor.fetchall()
    if not rows:
        print("No questions found for this case.")
        return
    
    # Group data by question
    questions_dict = {}
    for q_id, q_text, a_id, opt_text, is_correct in rows:
        if q_id not in questions_dict:
            questions_dict[q_id] = {
                "question_text": q_text,
                "options": []
            }
        questions_dict[q_id]["options"].append((a_id, opt_text, is_correct))
    
    # Gameplay
    score = 0
    for q_id, data in questions_dict.items():
        print(f"\n{data['question_text']}")
        options = data["options"]
        
        for i, (_, opt_text, _) in enumerate(options, start=1):
            print(f"  {i}. {opt_text}")
        
        try:
            choice = int(input("Choose your answer (1-4): "))
            if choice < 1 or choice > len(options):
                raise ValueError
        except ValueError:
            print("Invalid input, skipping question.")
            continue
        
        chosen = options[choice - 1]
        if chosen[2] == 1:
            print("Correct!")
            score += 1
        else:
            print("Wrong.")
    
    print("-" * 60)
    print(f"Final Score for Case {case_id}: {score}/{len(questions_dict)}")
    print("=" * 60)

# --- Main Menu ---
def main():
    print("=================================")
    print(" Welcome to the Detective Game ")
    print("=================================")
    show_cases()

# Run the game
main()
```
